```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
H = 10.
w = 1.
```

```python
pts = np.array([[2*w,1],
                [H-2*w,1],
                [H/2,H/2-w]])
```

```python
plt.fill(*pts.T);
```

```python
n = 10
G0x = np.linspace(pts[0,0], pts[1,0], n, endpoint=False)
G0y = np.linspace(pts[0,1], pts[1,1], n, endpoint=False)
```

```python
np.column_stack((G0x, G0y))
```

```python
def split(p0, p1, n):
    return np.column_stack((np.linspace(p0[0], p1[0], n, endpoint=False),
                            np.linspace(p0[1], p1[1], n, endpoint=False)))
```

```python
def split_border(pts, n):
    lst = []
    k = pts.shape[0]
    for i in range(k):
        lst.append(split(pts[i], pts[(i+1)%k], n))
    return np.vstack(tuple(lst))
```

```python
G = split_border(pts, n)
```

```python
plt.fill(*pts.T);
plt.plot(*G.T, '.k');
```

```python
K = 20
```

```python
lst = []
for i in range(pts1.shape[0]):
    lst.append(split(pts1[i,:2], pts1[i, 2:], K)[1:])
inpts = np.vstack(tuple(lst))
```
