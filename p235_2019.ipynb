````python
g = 0.25
````

````python
G0y = np.arange(0, 4, g)
G0x = np.zeros_like(G0y)
G0y, G0x
````

````python
G1x = np.arange(0, 10, g)
G1y = np.zeros_like(G1x)+4
G1x, G1y
````

````python
alpha = np.radians(np.arange(180, 270, 5))
G2x = 2*np.cos(alpha)+12
G2y = 2*np.sin(alpha)+4
````

````python
G3x = np.arange(12, 22, g)
G3y = np.zeros_like(G1x)+2
G3x, G3y
````

````python
G4y = np.arange(2, 0, -g)
G4x = np.zeros_like(G4y)+22
G4y, G4x
````

````python
G5x = np.arange(22, 0, -g)
G5y = np.zeros_like(G5x)
G5x, G5y
````

````python
GNodes = np.vstack((np.vstack((G0x, G0y)).T,
                    np.vstack((G1x, G1y)).T,
                    np.vstack((G2x, G2y)).T,
                    np.vstack((G3x, G3y)).T,
                    np.vstack((G4x, G4y)).T,
                    np.vstack((G5x, G5y)).T,
                   ))
````

````python
INodes = np.vstack((X.ravel(), Y.ravel())).T
````

````python
Nodes = np.vstack((GNodes, INodes))
````

````python
tri = Delaunay(Nodes)
````

````python
Elems = tri.simplices.copy()
````

````python
Elems
````

````python
Nodes.shape[0], Elems.shape[0], Elems.max(), Elems.min()
````

````python
plt.figure(figsize=(10,5))
plt.triplot(Nodes[:,0], Nodes[:,1], Elems)
plt.axis('equal')
````

````python
emask = np.logical_not(np.all(Elems < GNodes.shape[0], axis=1))
````

````python
Elems = Elems[emask]
````

````python
Elems.shape
````

````python
plt.figure(figsize=(10,5))
plt.triplot(Nodes[:,0], Nodes[:,1], Elems)
#plt.xlim(9.5, 12.5)
#plt.axis('equal')
````

````python
E = 2e7 # Н/см^2
mu = 0.25
t = 0.5 # см
P = 44e3 # Н/см^2
````

````python
Coords = Nodes[Elems]
````

````python
Coords.shape
````

````python
Nodes.shape
````

````python
A = 0.5*np.cross(Coords[:,1]-Coords[:,0], Coords[:,2]-Coords[:,0])
````

````python
b = np.roll(Coords[:,:,1], -1, axis=1) - np.roll(Coords[:,:,1], 1, axis=1)
b.shape
````

````python
c = np.roll(Coords[:,:,0], 1, axis=1) - np.roll(Coords[:,:,0], -1, axis=1)
c.shape
````

````python
Elems.shape, A.shape, b.shape, c.shape
````

```      
      _                      _
     | b0,  0, b1,  0, b2,  0 |
 B = |  0, c0,  0, c1,  0, c2 |*(1/2A)
     |_c0, b0, c1, b1, c2, b2_|
```

````python

B.shape
````

````python
B = np.zeros((Elems.shape[0],3,6))

B[:,0,0::2] = b
B[:,1,1::2] = c

B[:,2,1::2] = b
B[:,2,0::2] = c

B *= 1/(2*A[:,None,None])
````

````python
D = E/(1-mu**2)*np.array([[1, mu, 0],
                          [mu, 1, 0],
                          [0, 0, (1-mu)/2]])
````

````python
BT = np.transpose(B, axes=(0,2,1))
````

````python
Ke = BT@D@B
````

````python
EFT = 2*Elems[:,[0,0,1,1,2,2]] + np.array([0,1,0,1,0,1])
````

````python
EFT[0, None].T
````

````python
K = np.zeros((2*Nodes.shape[0],)*2)

#plus1 = np.array([0,1,0,1,0,1])
#plus1

for i, e in enumerate(Elems):
    #dof = 2*e[[0,0,1,1,2,2]] + plus1
    K[EFT[i,None].T, EFT[i]] += Ke[i]
````

````python
fixed = np.arange(17)
fixed = np.hstack((fixed*2, fixed*2+1))
symm = np.arange(122, GNodes.shape[0]+1)*2+1
force = np.arange(114, 123)*2
````

````python
plt.figure(figsize=(20,20))
plt.plot(GNodes[:,0], GNodes[:,1], '.')
for i, Node in enumerate(GNodes):
    plt.text(Node[0], Node[1], str(i))
````

````python
F = np.zeros(2*Nodes.shape[0])
F.shape
````

````python
# force
f = 0.5*P*t*(Nodes[114,1]-Nodes[115,1])
F[force] = f
F[force[1:-1]] += f
````

````python
# fixed
K[fixed,:] = 0
K[:,fixed] = 0
K[fixed, fixed] = 1
F[fixed] = 0
````

````python
# symm
K[symm,:] = 0
K[:,symm] = 0
K[symm, symm] = 1
F[symm] = 0
````

````python
# solve
U = np.linalg.solve(K, F)
````

````python
np.abs(U).max()
````

````python
rNodes = Nodes + 10000*U.reshape(-1,2)
````

````python
plt.figure(figsize=(10,5))
plt.triplot(rNodes[:,0], rNodes[:,1], Elems, color='r')
plt.plot(GNodes[:,0], GNodes[:,1], 'b')
plt.axis('equal')
````

````python
Eps = B@U[EFT, None]
Eps.shape
````

````python
Sig = D@Eps
Sig.shape
````

````python
# sigma0 = 0.5*(sig_x + sig_y) + sqrt((sig_x-sig_y)**2+tau_xy**2)
S0 = 0.5*(Sig[:,0]+Sig[:,1]) + \
     ((0.5*(Sig[:,0]-Sig[:,1]))**2+Sig[:,2]**2)**0.5
````

````python
# sigma0 = 0.5*(sig_x + sig_y) + sqrt((sig_x-sig_y)**2+tau_xy**2)
S1 = 0.5*(Sig[:,0]+Sig[:,1]) - \
     ((0.5*(Sig[:,0]-Sig[:,1]))**2+Sig[:,2]**2)**0.5
````
